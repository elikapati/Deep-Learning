Demonstration of onset diabetes within five years on pima indians checking the patient medical record.

- This is a simple application on pima indians diabetes data set to demonstrate neural network model structure. 
- It uses same data set for training, validation and testing.

- Demonstrate how to evaluate performance by splitting data into train, validate sets using 
  - automatic split
  - manual split
  - k-fold split

```python
from numpy import loadtxt
from tensorflow import keras
from tensorflow.keras import models, layers
```

```python
# Load dataset

dataset = loadtxt('Data/pima-indians-diabetes.csv', delimiter=',')
X = dataset[:, 0:8]
y = dataset[:, 8]
```

```python
# define model
model = models.Sequential()
model.add(layers.Dense(12, input_dim=8, activation='relu'))
model.add(layers.Dense(8, activation='relu'))
model.add(layers.Dense(1, activation='sigmoid'))
```

```python
model.compile(loss='binary_crossentropy', optimizer='adam', metrics=['accuracy'])
```

```python
model.fit(X, y, epochs=150, batch_size=10, verbose=1)
```

```python
# evaluate the keras model
_, accuracy = model.evaluate(X, y)
print('Accuracy: %.2f' % (accuracy*100))
```

```python
# make class predictions with the model
predictions = model.predict_classes(X)
```

```python
# summarize the first 5 cases
for i in range(5):
    print('%s => %d (expected %d)' % (X[i].tolist(), predictions[i], y[i]))
```

__How do we Evaluate model?__

Above, we have created simple model to demonstrate the structure. We have used same dataset for training, validation and testing, but it is not fair. So to evaluate the model, we seprate data into training, validation and testing sets and evaluate how best our model performs. Splitting of data can be done in one of 3 ways:
- a) automatic splitting: we specify %of data to be split into train and validate
- b) manual splitting: we split manually into train and validate sets
- c) k-fold splitting: we use library to do a rational and unbiased split into train and validate sets

__a) Automatic splitting__

```python
from numpy import loadtxt
from tensorflow import keras
from tensorflow.keras import models, layers

# Load dataset
dataset = loadtxt('Data/pima-indians-diabetes.csv', delimiter=',')
X = dataset[:, 0:8]
y = dataset[:, 8]

# define model
model = models.Sequential()
model.add(layers.Dense(12, input_dim=8, activation='relu'))
model.add(layers.Dense(8, activation='relu'))
model.add(layers.Dense(1, activation='sigmoid'))

model.compile(loss='binary_crossentropy', optimizer='adam', metrics=['accuracy'])

model.fit(X, y, validation_split=0.33, epochs=150, batch_size=10)

# evaluate the keras model
_, accuracy = model.evaluate(X, y)
print('Accuracy: %.2f' % (accuracy*100))
```

__b) Manual splitting__

```python
from numpy import loadtxt
from tensorflow import keras
from tensorflow.keras import models, layers
from sklearn.model_selection import train_test_split

# Load dataset
dataset = loadtxt('Data/pima-indians-diabetes.csv', delimiter=',')
X = dataset[:, 0:8]
y = dataset[:, 8]

# split into 67% for train and 33% for validation
X_train, X_valid, y_train, y_valid = train_test_split(X, y, test_size=0.33)

# define model
model = models.Sequential()
model.add(layers.Dense(12, input_dim=8, activation='relu'))
model.add(layers.Dense(8, activation='relu'))
model.add(layers.Dense(1, activation='sigmoid'))

model.compile(loss='binary_crossentropy', optimizer='adam', metrics=['accuracy'])

model.fit(X_train, y_train, validation_data=(X_valid, y_valid), epochs=150, batch_size=10)

# evaluate the keras model
_, accuracy = model.evaluate(X, y)
print('Accuracy: %.2f' % (accuracy*100))
```

__c) k-fold splitting__

```python
import numpy
from numpy import loadtxt
from tensorflow import keras
from tensorflow.keras import models, layers
from sklearn.model_selection import StratifiedKFold

# Load dataset
dataset = loadtxt('Data/pima-indians-diabetes.csv', delimiter=',')
X = dataset[:, 0:8]
y = dataset[:, 8]

# define 10-fold cross validation test harness
kfold = StratifiedKFold(n_splits=10, shuffle=True)

cvscores= []

for train, valid in kfold.split(X, y):
    # define model
    model = models.Sequential()
    model.add(layers.Dense(12, input_dim=8, activation='relu'))
    model.add(layers.Dense(8, activation='relu'))
    model.add(layers.Dense(1, activation='sigmoid'))

    model.compile(loss='binary_crossentropy', optimizer='adam', metrics=['accuracy'])

    model.fit(X[train], y[train], epochs=150, batch_size=10, verbose=0)
    
    # evaluate the keras model
    scores = model.evaluate(X[valid], y[valid], verbose=0)
    print('%s: Accuracy: %.2f' % (model.metrics_names[1], scores[1]*100))
    cvscores.append(scores[1]*100)
```

```python
print('Average mean and standard deviation of scores = %.2f%% (+/- %.2f%%)' % (numpy.mean(cvscores), numpy.std(cvscores)))
```
